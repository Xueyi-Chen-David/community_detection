```python
import os

# Include Training dataset
target_dir = os.getcwd()+"\Training"
circles_file = []

data = []

import re

target_dir = os.getcwd()+"\Training"

dirFiles = os.listdir(target_dir)
dirFiles.sort(key=lambda f: int(re.sub('\D', '', f))) # sorted all files

for filename in dirFiles:
    with open(os.path.join(target_dir, filename), 'r') as f:
        circles_file.append(os.path.join(target_dir, filename))
        data += f.readlines()

circle_data = []
for i in range(len(data)):
    circle_data.append([int(i) for i in data[i].split()[1:]]) # circle* is seems not nesseary

print(f"number of trainin data: {len(circle_data)}")

del target_dir, dirFiles
```

```python
egonets = []
egonet_file = []

target_dir = os.getcwd()+"\egonets"

dirFiles = os.listdir(target_dir)
dirFiles.sort(key=lambda f: int(re.sub('\D', '', f))) # sorted all files


for filename in dirFiles:
    with open(os.path.join(os.getcwd()+"\egonets", filename), 'r') as f:
        egonet_file.append(os.path.join(os.getcwd()+"\egonets", filename))
        egonets.append(f.readlines())
        
    
del target_dir, dirFiles
```

```python
import networkx as nx
from collections import defaultdict
import sys


def draw(egonet_file):
    G = nx.Graph()

    for line in open(egonet_file):
        e1, es = line.split(':')
        es = es.split()
        for e in es:
            if e == e1: continue
            G.add_edge(int(e1),int(e))
    return G
        
```

```python
G = draw(egonet_file[1])
pos=nx.spring_layout(G,scale=100, seed=3113794652)


nx.draw(G ,node_size=10, width=0.5,  with_labels = True, font_size=10, pos=pos)
```

```python
def try_int(s):
    try:
        return int(s)
    except ValueError:
        return s

all_egonets = []
    
for a in range(len(egonets)):
    temp = {}
    for i in range(len(egonets[a])):

        row = egonets[a][i].replace('\n', '').replace(':', '').split(' ')

        temp[int(row[0])] = [try_int(j) for j in row[1:]]
        
    all_egonets.append(temp)
    
new_egonets = []
for i in range(len(all_egonets)):
    temp = {list(all_egonets[i].keys())[0] - 1: list(all_egonets[i].keys())}
    temp.update(all_egonets[i])
    new_egonets.append(temp)
```
